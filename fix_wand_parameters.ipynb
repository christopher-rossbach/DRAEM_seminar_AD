```python
print("importing")
import wandb
from tqdm import tqdm
print("imported")

runs = wandb.Api().runs("team-cr/DRAEM_seminar_AD")
```

```python
changed = 0
for run in tqdm(runs):
    if "original_lr" in run.tags:
        print(f"Updating run: {run.name} ({run.id})")
        run.config["lr_scheduler"] = "multi_step_0.1_0.572_0.858"
        run.update()
        changed += 1
print(f"Total runs updated: {changed}")
```

```python
changed = 0
for run in tqdm(runs):
    if "amp" in run.config:
        run.config["amp_num"] = 1 if run.config["amp"] else 0
        run.update()
        changed += 1
print(f"Total runs updated: {changed}")
```

```python
# Get the latest run (first one from the runs)
latest_run = runs[0]

print(f"Latest run: {latest_run.name} ({latest_run.id})")
print(f"Status: {latest_run.state}")
print("\n" + "="*50)
print("CONFIG FIELDS:")
print("="*50)
for key, value in latest_run.config.items():
    print(f"{key}: {value}")

print("\n" + "="*50)
print("SUMMARY FIELDS:")
print("="*50)
for key, value in latest_run.summary.items():
    print(f"{key}: {value}")

print("\n" + "="*50)
print("HISTORY KEYS (available metrics):")
print("="*50)
run_history = latest_run.history()
if len(run_history) > 0:
    if isinstance(run_history, list):
        print(list(run_history[0].keys()))
    else:
        print(list(run_history.columns))
else:
    print("No history available")
```
