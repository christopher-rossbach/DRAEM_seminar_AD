```python
print("importing")
import wandb
from tqdm import tqdm
print("imported")

runs = wandb.Api().runs("team-cr/DRAEM_seminar_AD")
```

```python
changed = 0
for run in tqdm(runs):
    if "orig_imp" in run.tags:
        print(f"Updating run: {run.name} ({run.id})")
        run.config["micro_batch_size"] = 8
        run.update()
        changed += 1
print(f"Total runs updated: {changed}")
```

```python
changed = 0
for run in tqdm(runs):
    if any(tag in run.tags for tag in ["micro_batch_intro", "orig_imp"]):
        print(f"Updating run: {run.name} ({run.id})")
        run.config["lr_scheduler"] = "multi_step_0.2_0.8_0.9"
        run.update()
        changed += 1
print(f"Total runs updated: {changed}")

changed = 0
for run in tqdm(runs):
    if not any(tag in run.tags for tag in ["micro_batch_intro", "orig_imp"]):
        print(f"Updating run: {run.name} ({run.id})")
        run.config["lr_scheduler"] = "cosine_annealing"
        run.update()
        changed += 1
print(f"Total runs updated: {changed}")
```
